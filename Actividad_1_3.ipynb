```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel('Inmuebles.xlsx')
df.head()
```

```python
#Eliminar la fila “0” de  “Unnamed” y Eliminar la columna “0” y la columna  “Referencia

df.columns = df.iloc[0]

# Eliminar la primera fila
df = df[1:]

# Restablecer el índice del DataFrame
df = df.reset_index(drop=True)

# Eliminar la primera columna
df = df.iloc[:, 1:]

df.head()
```

```python
df.to_csv('Filtro1.csv')
#convertir archivo filtrado

from google.colab import files
files.download('Filtro1.csv')
```

```python
#Todos las registros de “Fecha Alta : igual a 2007-04-29 00:00:00 y a 2004-01-02 00:00:00

# Convertir la columna 'Fecha Alta' a tipo datetime
df['Fecha_Alta'] = pd.to_datetime(df['Fecha_Alta'])

# Filtrar registros por las fechas especificadas (usando | para "o")
filtro2 = df[(df['Fecha_Alta'] == '2007-04-29 00:00:00') | (df['Fecha_Alta'] == '2004-01-02 00:00:00')]

filtro2
```

```python
filtro2.to_csv('Filtro2.csv')
#convertir archivo filtrado

from google.colab import files
files.download('Filtro2.csv')
```

```python
# Todos los registros de la columna Tipo que sean = Parking, Local, Industrial y Piso”

tipos = ['Parking', 'Local', 'Industrial', 'Piso']

filtro3 = df[df['Tipo'].isin(tipos)]

filtro3
```

```python
filtro3.to_csv('Filtro3.csv')
#convertir archivo filtrado

from google.colab import files
files.download('Filtro3.csv')
```

```python
#Todos los registros con “Provincia” igual a “Girona” y con “Superficie” mayor a “100”

filtro4 = df[(df['Provincia'] == 'Girona') & (df['Superficie'] > 100)]

# Mostrar todos los registros filtrados
filtro4
```

```python
filtro4.to_csv('Filtro4.csv')
#convertir archivo filtrado

from google.colab import files
files.download('Filtro4.csv')
```

```python
#Todos los registros con “Precio Venta: mayor a 1000,000” o “vendedor” igual a  “Carmen o Joaquin”
filtro5 = df[(df['Precio Venta'] > 1000000) | (df['Vendedor'].isin(['Carmen', 'Joaquín']))]
filtro5
```

```python
filtro5.to_csv('Filtro5.csv')
#convertir archivo filtrado

from google.colab import files
files.download('Filtro5.csv')
```

```python
#Todos los registros que no sean de “Provincia : Barcelona“
Filtro6=df[df['Provincia']=='Barcelona']
Filtro6
```

```python
Filtro6.to_csv('Filtro6.csv')
#convertir archivo filtrado

from google.colab import files
files.download('Filtro6.csv')
```

```python
#Todos los registros que no sean de “Operación: alquiler “ ni “Tipo: oficina“

filtro7 = df[~((df['Operación'] == 'Alquiler') & (df['Tipo'] == 'Oficina'))]

filtro7
```

```python
filtro7.to_csv('Filtro7.csv')
#convertir archivo filtrado

from google.colab import files
files.download('Filtro7.csv')
```
